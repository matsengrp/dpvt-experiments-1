```python
def hamming_distance(seq1, seq2):
    # Check if the lengths of the sequences are equal
    if len(seq1) != len(seq2):
        raise ValueError("Sequences must be of equal length to compute Hamming distance.")
    
    # Count the number of positions where elements are different
    distance = sum(el1 != el2 for el1, el2 in zip(seq1, seq2))
    
    return distance

def mutations_per_branch(tree):
    # Return number of branches (values) with x mutations (keys)
    output = {}
    for node in tree.traverse("preorder"):
        if node.up != None:
            num_mutations = hamming_distance(node.sequence, node.up.sequence)
            if num_mutations in output:
                output[num_mutations] += 1
            else:
                output[num_mutations] = 1
    if 0 in output:
        num_nodes = len([node for node in tree.traverse() if node.up != None])
        fraction_no_mutation_branches = output[0]/num_nodes
    else:
        fraction_no_mutation_branches = 0
    return output, fraction_no_mutation_branches

from collections import defaultdict

def merge_and_sum_dicts(dict_list):
    # Create a defaultdict to hold the sums
    merged_dict = defaultdict(int)
    
    # Iterate through each dictionary in the list
    for d in dict_list:
        for key, value in d.items():
            merged_dict[key] += value
    
    # Convert defaultdict back to a regular dictionary
    return dict(merged_dict)
```

```python
with open(filename, "rb") as f:
    data = pickle.load(f)

dict_list = []
all_fractions_no_mutation_branches = []
for tree in data:
    d, fraction_no_mutation_branches = mutations_per_branch(tree)
    dict_list.append(d)
    all_fractions_no_mutation_branches.append(fraction_no_mutation_branches)

all_mutations = merge_and_sum_dicts(dict_list)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(list(all_mutations.items()), columns = ["mutations_per_branch", "count"])
print(df["mutations_per_branch"].max())
sns.scatterplot(data=df, x="mutations_per_branch", y="count")
plt.savefig("./data_props.pdf")
plt.clf()

```

# Create combined training/testing set of 30LeafPerfect and 10LeafPerfect

```python
import pickle

with open("10leaf_perfect_distinct_trees_train.p", "rb") as f:
    TenLeafTrain = pickle.load(f)
with open("30leaf_perfect_distinct_trees_train.p", "rb") as f:
    ThirtyLeafTrain = pickle.load(f)

with open("10leaf_perfect_distinct_trees_test.p", "rb") as f:
    TenLeafTest = pickle.load(f)
with open("30leaf_perfect_distinct_trees_test.p", "rb") as f:
    ThirtyLeafTest = pickle.load(f)
```
