```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

current_date = datetime.datetime.now().strftime("%Y-%m-%d")

property_csv = "../larch-data-small/data_properties_harrington-small_2024-06-10.csv"
data_name = "harrington-small"

df=pd.read_csv(property_csv)
print(df.columns)
df["non_MP_edge_frac"] = df["non MP edges"] / (df["MP edges"] + df["non MP edges"]) # fraction of non-MP edges
plt.clf()
sns.scatterplot(data=df, x="num_leaves", y="num_trees", hue="non_MP_edge_frac")
plt.savefig("plots/"+data_name+"_"+current_date+"_num_leaves_vs_trees.pdf")
plt.clf()
sns.scatterplot(data=df, x="num_leaves", y="alignment_length", hue="non_MP_edge_frac")
plt.savefig("plots/"+data_name+"_"+current_date+"_num_leaves_vs_sites.pdf")
plt.clf()
sns.scatterplot(data=df, x="alignment_length", y="num_trees", hue="non_MP_edge_frac")
plt.savefig("plots/"+data_name+"_"+current_date+"_num_sites_vs_trees.pdf")
print(sum(df["MP edges"]))
print(sum(df["non MP edges"]))
print(df)
print(max(df["num_leaves"]))
```

```python
import os
import csv
import pandas as pd
import numpy as np

# Directory containing subdirectories with "larch-output.pb" files
root_dir = "../../data-copy/"

# Iterate through subdirectories
for subdir in os.listdir(root_dir):
    subdir_path = os.path.join(root_dir, subdir)
    if os.path.isdir(subdir_path):
        print(subdir_path)
        output_file_path = os.path.join(subdir_path, "num_children_dag_trees.csv")
        if os.path.isfile(output_file_path):
            # slightly complicated way to read csv, because number of entries per row depends
            # on number of nodes in tree, which varies because of multifurcations
            data = []
            with open(output_file_path, 'r') as file:
                reader = csv.reader(file)
                for row in reader:
                    data.append(row)

            # Determine the maximum number of columns
            max_cols = max(len(row) for row in data)

            # Fill up rows with too few entries with NaN values
            for row in data:
                while len(row) < max_cols:
                    row.append(None)

            # Create a DataFrame from the processed data
            df = pd.DataFrame(data)
            df = df.apply(pd.to_numeric, errors='coerce')
            counts_df = df.apply(lambda row: row.value_counts(), axis=1)
            print(counts_df)
            df['median'] = df.median(axis=1, skipna=True)
            df['mean'] = df.mean(axis=1, skipna=True)
            df['max'] = df.max(axis=1, skipna=True)
            # print(output_file_path)
            # print(df[['max', 'mean', 'median']])
```
