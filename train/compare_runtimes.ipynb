```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tbparse
from IPython.display import display
%matplotlib inline
```

```python
loop_csv = "_output/run.cpu_TODAY_pp_dpvt_with_site_loop/result_csvs_and_pdfs/summary.csv"
non_loop_csv = "_output/run.cpu_TODAY/result_csvs_and_pdfs/summary.csv"
loop_df = pd.read_csv(loop_csv)
non_loop_df = pd.read_csv(non_loop_csv)
```

```python
for df in [loop_df, non_loop_df]:
    df['model_and_train_data'] = df['model'] + "-" + df['train_data']
    df['label'] = df[['model','train_data','test_data', 'param_id']].apply(lambda x:'\n'.join(x), axis=1)
    df['num_train_leaves'] = df['train_data'].str.split("leaf").str[0].astype(int)
loop_df['train_clog_path'].replace("TODAY", "TODAY_pp_dpvt_with_site_loop")
model_list = loop_df['model'].unique()
num_train_leaf_list = loop_df['num_train_leaves'].unique()
```

```python
# get dataframe from lightning log

def get_df_from_log(log_path):
    reader = tbparse.SummaryReader(log_path)
    return reader.scalars
```

```python
# individual bar plots
x_measures = {
  "Training Runtime (in secs)": "train_walltime",
}

loop_df["source"] = "loop_df"
non_loop_df["source"] = "non_loop_df"

df = pd.concat([loop_df, non_loop_df], ignore_index=True)
df["model_and_num_train_trees"] = df["model"] + "-" + df["train_data"].str.split("trees").str[0].str.split("_").str[-1]

display(df)

for x_title,x_col in x_measures.items():
  fig, axes = plt.subplots(nrows=1, ncols=len(model_list), figsize=(15, 6), sharey=True)
  x_min = df[x_measures[x_title]].min() * 0.99
  x_max = df[x_measures[x_title]].max() * 1.01
  for ax, num_train_leaves in zip(axes, num_train_leaf_list):
      this_df = df[df['num_train_leaves']==num_train_leaves]
      sns.barplot(y="model_and_num_train_trees", x=f"{x_col}", data=this_df, palette='deep', ax=ax, hue="source")
      ax.set_xlabel(f"{x_title}")
      ax.set_title(f"num_train_leaves {num_train_leaves}")
      # ax.set_xlim(x_min, x_max)
      plt.tight_layout()
  plt.savefig(f'plots/runtime_comparison_site_loop_vs_no_loop.barplot.pdf')
  plt.show()
  plt.clf()
```

```python
# Pivot the data so we have separate columns for df1 and df2 values
pivot_df = df.pivot(index=["label", "model", "model_and_train_data"], columns="source", values="test_auroc").reset_index()
pivot_df["label"] = pivot_df["label"].str.replace("\n", "")


# Compute the difference
pivot_df["diff"] = pivot_df["loop_df"] - pivot_df["non_loop_df"]

# # Sort by difference for better readability
pivot_df = pivot_df.sort_values("model_and_train_data", ascending=False)


for model in model_list:
    # Plot
    plt.figure(figsize=(15, 6))
    sns.barplot(data=pivot_df[pivot_df["model"]==model], y="label", x="diff", palette="coolwarm")

    plt.axvline(0, color="black", linestyle="--")  # Reference line for no difference
    plt.xlabel("Difference (loop_df - non_loop_df)")
    plt.ylabel("Label")
    plt.title("Difference in test_auroc Between non_loop_df and loop_df")
    plt.show()

```
