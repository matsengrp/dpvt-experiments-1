```python
import os
from Bio import SeqIO
import pandas as pd

def count_taxa_in_fasta(filepath):
    """Count the number of taxa in a FASTA file."""
    count = 0
    try:
        with open(filepath, 'r') as file:
            for record in SeqIO.parse(file, "fasta"):
                count += 1
    except Exception as e:
        print(f"Error reading {filepath}: {e}")
    return count

def find_directories_with_fasta_in_range(root_dir, min_taxa, max_taxa):
    """Find directories under root_dir containing FASTA files with taxa count in the specified range."""
    qualifying_subdirs = []
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('input.fasta'):
                filepath = os.path.join(subdir, file)
                taxa_count = count_taxa_in_fasta(filepath)
                if min_taxa <= taxa_count <= max_taxa:
                    qualifying_subdirs.append(os.path.basename(subdir))
                    break  # Stop checking if one file in the directory qualifies
    return qualifying_subdirs

def write_directories_to_csv(directories, csv_file_path):
    """Write the list of directory names to a CSV file using pandas."""
    df = pd.DataFrame(directories, columns=['Directory Name'])
    df.to_csv(csv_file_path, index=False)

# Example usage:
root_directory = 'larch-data-small'
min_taxa = 0
max_taxa = 40
qualifying_dirs = find_directories_with_fasta_in_range(root_directory, min_taxa, max_taxa)
print("Subdirectories with 0-50 taxa in FASTA files:")
for dir in qualifying_dirs:
    print(dir)
write_directories_to_csv(qualifying_dirs, root_directory + "/alignments-with-" + str(min_taxa) + "-to-" + str(max_taxa) + "-taxa.csv")
```
