```python
import pickle
import os
import pandas as pd
```

```python
directory = "/fh/fast/matsen_e/shared/dpvt/datasets/"

data_list = []

for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    if os.path.isfile(file_path) and "alisim_alignment" in file_path:
        with open(file_path, "rb") as f:
            data = pickle.load(f)
            avg_sequence_length = sum([len(i.sequence) for i in data])/len(data)
            avg_num_sequences = sum([len(i) for i in data])/len(data)
            data_list.append([filename, avg_num_sequences, avg_sequence_length, len(data)])

df = pd.DataFrame(data_list, columns=["name", "avg_num_sequences", "avg_num_sites", "num_trees"])
df = df.sort_values(by = "name")
df
```
