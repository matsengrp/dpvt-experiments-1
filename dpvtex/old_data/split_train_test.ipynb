```python
import pickle
from sklearn.model_selection import train_test_split
from collections import Counter
import pandas as pd
```

```python
filename = "4leaf4site.p"
dataname = filename[:-2]
```

```python
with open(filename, "rb") as f:
    tree_dict = pickle.load(f)

trees = list(tree_dict.keys())
labels = list(tree_dict.values())


# For Stratifying
sum_of_ones = [sum(label) for label in labels]
counter = Counter(sum_of_ones)

# Convert sums to a categorical variable for balancing number of non-MP edges in train/test/val
categories = pd.qcut(
    sum_of_ones, q=min(len(counter), 4), labels=False, duplicates="drop"
    )

train_trees, test_trees, train_labels, test_labels = train_test_split(trees, labels, train_size = 0.8, stratify = categories)

train_data_name = dataname + "_train.p"
test_data_name = dataname + "_test.p"

train_dict = {i:j for (i,j) in zip(train_trees, train_labels)}
test_dict = {i:j for (i,j) in zip(test_trees, test_labels)}

with open(train_data_name, "wb") as f:
    pickle.dump(train_dict, f)

with open(test_data_name, "wb") as f:
    pickle.dump(test_dict, f)
```
